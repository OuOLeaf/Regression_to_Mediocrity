```python
# packages
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML

# distribution setting
p = 0.5
bins = 10
times = 20
beads = 10000
add_beads = int(beads/times)

# pre-sample all data
np.random.seed(0)
samples = list(np.random.binomial(n = bins, p = p, size = beads))


# initialization function: histogram details
def update_hist(num, data):
    n = add_beads * (num+1)
    plt.cla()
    plt.hist(data[:n], bins = list(range(1, 11)), color = "orange", edgecolor = "purple", alpha = 1)
    plt.suptitle("Galton Board")
    plt.title(f"Binomial Simulation of {n} samples")
    plt.ylim(0, 2600)

plt.style.use('dark_background')
fig = plt.figure()


# call the animator. frame：num for loop, interval：time space
anim = animation.FuncAnimation(fig, update_hist,
                               frames= 20, interval=400, fargs=(samples, ))

# display
HTML(anim.to_html5_video())
```
