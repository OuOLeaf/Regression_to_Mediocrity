```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


p = 0.5
bins = 10
beads = 10000
np.random.seed(0)
samples = np.random.binomial(n = bins, p = p, size = beads)


plt.style.use('dark_background')
plt.hist(samples, bins = list(range(1, 11)), color = "orange", edgecolor = "purple", alpha = 0.8)
plt.title("Galton Binomial Simulation")
plt.show()
```
